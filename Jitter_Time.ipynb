# Calculating Jitter as a function of Time

Using the template we formed in the previous notebook, we now find the jitter of the data as a function of total subintegration time

```python
import os
import psrchive
import numpy as np
import matplotlib.pyplot as plt
from uncertainties import ufloat
from uncertainties import unumpy
```

Collapse 2 subints of the data repeatedly to obtain new files with different subintegration time

```python
split = 8
```

Load the template file for jitter analysis

```python
template = psrchive.Archive_load("fits/template.fits")
template = template.get_data()
```

Load observed data and collapse it to a single frequency

```python
data = psrchive.Archive_load("fits/493_446.fits")
data.fscrunch()
data.unload("fits/time/493_1.fits")
```

Increasing subintegration time by a factor of 2 repeatedly and saving the resulting files

```python
values = [data.get_nsubint()]
for i in range(split - 1):
    data.tscrunch(2)
    x = data.get_nsubint()
    values.append(x)
    y = data.get_nchan()
    name = f"fits/time/{x}_{y}.fits"
    data.unload(name)

print(values)
```

```python
values = [493, 247, 124, 62, 31, 16, 8, 4]
```

The ArrivalTime class calls the 'pat' module to make a timing file by using the template file

```python
arrtim = psrchive.ArrivalTime()
arrtim.set_shift_estimator("PGS")  # can change to FDM (MCMC), COF, GIS
arrtim.set_format("Tempo2")

std = psrchive.Archive_load("fits/template.fits")
arrtim.set_standard(std)
```

Saving the Time of Arrivals for different subintegration times

```python
toas = []

for i in range(split):
    obs = psrchive.Archive_load(f"fits/time/{values[i]}_1.fits")
    arrtim.set_observation(obs)

    toa = arrtim.get_toas()
    toas.append(toa)
```

Writing the Time of Arrivals to a .tim file for each subintegration time

```python
for i in range(split):
    f = open(f"tim/time/{values[i]}.tim", "w")
    f.write("FORMAT 1\n")
    for j in range(len(toas[i])):
        f.write(toas[i][j] + "\n")
f.close()
```

Once we have the timing file, we use 'tempo2' to calculate its pre-fit residuals

```python
residuals = []

for i in range(len(values)):
    residual = []
    name = values[i]
    ! tempo2 -f J1136+1551.par tim/time/{name}.tim -residuals
    f = open("residuals.dat", "r")
    for line in f:
        residual.append([float(line.split()[1]), float(line.split()[2])])
    residuals.append(residual)
```

Plotting pre-fit residuals against time. We don't see many outliers in the data, but there are some point with large error bars which need to be removed

```python
i = 0
xvalues = np.arange(len(residuals[i]))
yvalues = np.array(residuals[i])[:,0]
yerr = np.array(residuals[i])[:,1]

slope, intercept = np.polyfit(xvalues, yvalues, 1)
yline = intercept + slope*xvalues

plt.errorbar(xvalues, yvalues, yerr, fmt=".")
plt.plot(xvalues, yline)
plt.xlabel("Subints")
plt.ylabel("Residuals (s)")
```

We filter out these outliers using MAD (median absolute deviation). Any point which is more than 3 MADs away from the median is removed

```python
valids = []

for i in range(len(residuals)):
    temp = np.array(residuals[i])
    valid = [1]*len(temp)
    timings = temp[:,0]
    
    mad = np.median(np.absolute(timings - np.median(timings)))
    # slope, intercept = np.polyfit(np.arange(len(temp)), timings, 1)
    x = np.median(timings)

    # stdev = 1.4826*MAD, 3*MAD = 2*stdev = 95% of data
    for k in range(len(timings)):
        # y = intercept + slope*k
        if abs(x-timings[k]) > 3*mad:
            valid[k] = 0
    valids.append(valid)

print([valids[i].count(1) for i in range(len(valids))])  
```

Similarly we also remove points with high error bars using MAD. Any point with an error bar more than 4.5 MADs away from the median is removed

```python
for i in range(len(residuals)):
    temp = np.array(residuals[i])
    timings = temp[:,1]

    mad = np.median(np.absolute(timings - np.median(timings)))
    x = np.median(timings)

    # stdev = 1.4826*MAD, 4.5*MAD = 3*stdev = 99.7% of data
    for k in range(len(timings)):
        if abs(x-timings[k]) > 4.5*mad:
            valids[i][k] = 0

print([valids[i].count(1) for i in range(len(valids))])
```

Plotting the filtered data

```python
i = 0
tim = residuals[i]
index = [j for j in range(len(tim)) if valids[i][j] == 1]

xvalues = np.arange(len(tim))[index]
yvalues = np.array(tim)[:,0][index]
yerr = np.array(tim)[:,1][index]

slope, intercept = np.polyfit(xvalues, yvalues, 1)
yline = intercept + slope*xvalues

plt.errorbar(xvalues, yvalues, yerr, fmt=".")
plt.plot(xvalues, yline)
plt.xlabel("Subints")
plt.ylabel("Residuals (s)")
```

Saving filtered data to new files. These will now be used to fit parameters

```python
for i in range(split):
    f = open(f"tim/time/{values[i]}_clean.tim", "w")
    f.write("FORMAT 1\n")
    for j in range(len(toas[i])):
        if valids[i][j] == 1:
            f.write(toas[i][j] + "\n")
f.close()
```

Fitting frequency of the pulsar using 'tempo2' for the filtered data. The ephemeris file gives the initial estimates for the parameters

```python
for i in range(split):
    name = values[i]
    ! tempo2 -f J1136+1551.par tim/time/{name}_clean.tim -fit F0 -residuals
    path = f"tim/time/{name}_residuals.dat"
    os.rename("residuals.dat", path)
```

The post fit residuals obtained from fitting have non zero RMS noise. This rms is the sum of jitter noise and radiometer noise. To find jitter, we subtract radiometer noise from the net RMS noise. We also keep track of the error bars while doing this.

First we calculate net RMS noise of the post fit residuals

```python
rms = []

for i in range(split):
    residual_value = []
    residual_error = []

    f = open(f"tim/time/{values[i]}_residuals.dat", "r")
    for line in f:
        line = line.split()
        residual_value.append(float(line[1]))
        residual_error.append(float(line[2]))

    residual_value = np.array(residual_value)
    residual_error = np.array(residual_error)
   
    residuals = unumpy.uarray(residual_value, residual_error)
    
    for j in range(len(residuals)):
        residuals[j] = residuals[j]**2
    
    rms.append(residuals.sum()/len(residuals))

print(rms)
```

Then we find radiometer noise by finding the RMS of the error bars of the data

```python
radiometer = []
for i in range(split):
    residual_error = []

    f = open(f"tim/time/{values[i]}_residuals.dat", "r")
    for line in f:
        line = line.split()
        residual_error.append(float(line[2]))

    residual_error = np.array(residual_error)
    mean = residual_error.sum()/len(residual_error)

    sum_new = 0
    for i in range(len(residual_error)):
        sum_new += (residual_error[i] - mean)**2
    sum_new /= len(residual_error)
    sum_new = sum_new**0.5
    
    x = ufloat(mean, sum_new)
    radiometer.append(x**2)

print(radiometer)
```

Finally, we calculate jitter by subtracting the radiometer noise from net RMS noise

```python
jitter = []
for i in range(split):
    x = (rms[i] - radiometer[i])**0.5
    jitter.append(x)

print(jitter)
```

Plotting jitter with error bars as a function of subintegration time

```python
xvalues = [3, 6, 12, 24, 48, 96, 192, 384]
yvalues = []
yerr = []
for i in range(split):
    yvalues.append((jitter[i].nominal_value)*1e6)
    yerr.append((jitter[i].std_dev)*1e6)

f = open("tim/time/1.txt", "w")
f.write("Subint (s)       Jitter (us)        Jitter Error (us)\n")
for i in range(8):
    value = f"{xvalues[i]} {yvalues[i]} {yerr[i]} \n"
    f.write(value)
f.close()

plt.errorbar(xvalues[:-1], yvalues[:-1], yerr=yerr[:-1], fmt=".--")
plt.xlabel("Subintegration Time (s)")
plt.ylabel("Jitter (us)")
```
